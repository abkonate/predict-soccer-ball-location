Function for calculating mean of players x,y positions on the pitch

```python
def caluclate_avrg_pos(minutes_step, pos_mean, player, pos_count, activity_count):
    avrg_pos_count = 0 
    for minutes in range(0,minutes_step):
        pos_mean[player][0] += pos_count[player][0][minutes]
        pos_mean[player][1] += pos_count[player][1][minutes]
        avrg_pos_count += activity_count[player][minutes] 
    if avrg_pos_count!=0:
        pos_mean[player][0] /= avrg_pos_count
        pos_mean[player][1] /= avrg_pos_count
    else:
        pos_mean[player][0] = 0 
        pos_mean[player][1] = 0
```

```python
def caluclate_activity_count(minutes_step, player, home_activity):
    
    activity_count_each = 0 
    for minutes in range(0, minutes_step):
        activity_count_each += home_activity[player][minutes] 
    
    if activity_count_each != 0:
        return 1
    else:
        return 0
```
