```python
from scipy.spatial import procrustes
import numpy as np
from matplotlib import style
import matplotlib.pyplot as plt

style.use('ggplot')





sample_position = np.array([[25, 7], [17.5, 20], [17.5, 47], [25, 61], [69, 7], [61, 20], 
              [61, 47], [69, 61], [75, 34], [92.5, 34]])


example = np.array([[38.64, 40.69], [39.45, 27], [54.90, 54.49], [52.37, 18.85], [51.97, 50.42], [49.05, 34.05], 
              [69.70, 36.65], [42.97, 53.14], [42.23, 16.06], [62.38, 33.30]])

plt.title(r'Sample position: 4-5-1')
for i in range(0, len(sample_position)):
    plt.scatter(sample_position[i][0], sample_position[i][1], marker="o", color="r", linewidths=0.5)
plt.show()


plt.title(r'BJK avrg position, Min:35')
for i in range(0, len(example)):
    plt.scatter(example[i][0], example[i][1], marker="o", color="k", linewidths=0.5)
plt.show()

mtx1, mtx2, disparity = procrustes(sample_position, example)


print('\n After fitting parameters')


plt.title(r'After fitting: Sample position, 4-5-1')
for i in range(0, len(mtx1)):
    plt.scatter(mtx1[i][0], mtx1[i][1], marker="o", color="r", linewidths=0.5)
plt.show()


plt.title(r'After fitting: BJK avrg position, Min:35')
for i in range(0, len(mtx2)):
    plt.scatter(mtx2[i][0], mtx2[i][1], marker="o", color="k", linewidths=0.5)
plt.show()
```
